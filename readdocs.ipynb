```python
from bs4 import BeautifulSoup
import codecs
text_file = codecs.open("data/TREC8all/Adhoc/fbis/fb396073", "r", "iso-8859-1").read()
soup = BeautifulSoup(text_file,'lxml')
```

```python
import en_core_web_md
import spacy
from spacy.symbols import ORTH, LEMMA, POS

nlp = spacy.load("en")
doc = nlp(u'This is a sentence.')
```

```python
d = {}
for doc in soup.find_all("doc"):
    strdoc = doc.docno.string.strip()
    print(strdoc)
    tokens = nlp(doc.text)
    for token in tokens:
        if not token.is_stop and token.is_alpha:
            #print(token.is_stop())
            #if token.is_alpha() is False:
            #    break
            strtok = token.lemma_.strip()
            if strtok not in d.keys():
                d[strtok] = {strdoc}
            elif strtok in d.keys():
                d[strtok].add(strdoc)
print("done")
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(d)
```
